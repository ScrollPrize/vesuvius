In this tutorial, we will explore how to load instance annotated cubes in a python script.

### Getting Started:
First, we need to import the Vesuvius package, which will help us quickly fetch the cube data from the repository.

```python
!pip install vesuvius
!vesuvius.accept_terms --yes
import vesuvius
from vesuvius import Cube
```

Let's examine which cubes are available on the data repository


```python
print(vesuvius.cubes())
```

What we observe is a dictionary indexed with scroll number, beam energy used for the scan, resolution (in um) and cube indices with z_y_x representing the origin of the reference frame.

Let us access to an available instance annotated cube.

```python
instance = Cube(scroll_id=1, energy=54, resolution=7.91, z=0, y=2408, x=4560)
```

Instead of accessing the data in chunks, since a cube already corresponds to a small region of the original scroll volume, the `Cube` class will download the full data directly.

```python
volume, mask = instance[:,:,:] # load the data
```

Let us plot a slice from both the raw volume and the annotated mask.

```python
from matplotlib import pyplot as plt
```

```python
slice_z = 128 # for instance z = 128
```

```python
# Displaying the 129th slice (index 128) side by side
plt.figure(figsize=(12, 6))

# Plot the volume image
plt.subplot(1, 2, 1)
plt.imshow(volume[slice_z], cmap='gray')
plt.title(f'Volume Slice {slice_z}')
plt.axis('off')  # Hide the axis

# Plot the mask image
plt.subplot(1, 2, 2)
plt.imshow(mask[slice_z], cmap='tab20b')
plt.title(f'Mask Slice {slice_z}')
plt.axis('off')  # Hide the axis

plt.show()
```

Fantastic! We are visualizing different instances annotated with different labels (colors)!
